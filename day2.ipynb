```python
# kendi datasetimizi olusturacagiz

s1={"characters":["Obi-Wan Kenobi","Han Solo","Leia Organa","Luke Skywalker","Darth Vader","Anakin Skywalker","Count Dooku"],"sword":["Blue","-","-","Green","Red","Blue","Red"]}
s2={"characters":["Obi-Wan Kenobi","Han Solo","Leia Organa","Luke Skywalker","Darth Vader","Anakin Skywalker","Count Dooku"],"connection":["Republican","Rebels","Rebels","Rebels","Empire","Republican","Rebels"]}
s3={"characters":["Obi-Wan Kenobi","Han Solo","Leia Organa","Luke Skywalker","Darth Vader","Anakin Skywalker","Count Dooku"],"age":[38,29,19,19,42,22,83]}
s4={"characters":["Obi-Wan Kenobi","Han Solo","Leia Organa","Luke Skywalker","Darth Vader","Anakin Skywalker","Count Dooku"],"height":[180,174,165,175,203,183,185]}
s5={"connection":["Republican","Separatist","Rebels","Empire"],"master":["Yoda","Palpatine","Mon Mothma","Palpatine"]}
s6={"connection":["Republican","Separatist","Rebels","Empire"],"side":["Good","Bad","Good","Bad"]}
```

```python
d1=pd.DataFrame(s1)
d2=pd.DataFrame(s2)
d3=pd.DataFrame(s3)
d4=pd.DataFrame(s4)
d5=pd.DataFrame(s5)
d6=pd.DataFrame(s6)
```

```python
df_m1=pd.merge(d1,d2)
df_m2=pd.merge(df_m1,d3)
df_m3=pd.merge(df_m2,d4)
df_m4=pd.merge(df_m3,d5)
df_sw=pd.merge(df_m4,d6)
```

```python
df_sw
```

```python
df_sw.info()
```

```python
categorical_columns = df_sw.select_dtypes(["object"]).columns
```

```python
for col in categorical_columns:
  df_sw[col] = df_sw[col].astype("category")
```

```python
(sns.FacetGrid(df_sw,
          hue="master",
          height=5)
.map(sns.kdeplot,"age",shade=True)
.add_legend())
plt.title('Effect of master on age distribution')
```
